## **Question1: Student Grades Processing**

```python

# You need to write a Python program to process a student's grades. The program needs to perform the following functions:
# Store information and notes for 10 students using a dictionary. 
# Let each student have their name, surname and grades (Midterm, Final and Oral grades). For example:

students = {
    "ahmet yilmaz" : [85,90,78],  # Midterm: 85, Final: 90, Oral: 78
    "Mehmet Demir" : [92,88,76],  # Midterm: 85, Final: 90, Oral: 78
    "Ayse Kaya"    : [78,89,95],  # Midterm: 85, Final: 90, Oral: 78
    "Zeynep Celik" : [65,70,80],  # Midterm: 85, Final: 90, Oral: 78
    "Ali Kara"     : [50,60,55],  # Midterm: 85, Final: 90, Oral: 78
    "Fatma Yildiz" : [88,85,90],  # Midterm: 85, Final: 90, Oral: 78
    "Murat Aydin"  : [72,68,74],  # Midterm: 85, Final: 90, Oral: 78
    "Elif Aksoy"   : [95,90,88],  # Midterm: 85, Final: 90, Oral: 78
    "Hakan Ozyurek": [45,50,55],  # Midterm: 85, Final: 90, Oral: 78
    "ahmet yilmaz" : [80,75,82]   # Midterm: 85, Final: 90, Oral: 78

}
```

### **•1-Calculate each student's GPA and add it to the dictionary.•**

```python

gpa_dict = {i: "{:.2f}".format(sum(students[i])/3) for i in students.keys()}
print(gpa_dict)
```

### **•2-Find the student with the highest GPA and print it on the screen.•**

```python
for key, value in gpa_dict.items():
    if value == max(gpa_dict.values()):
        print(key)
```

### **•3- Separate each student's name from their surname and store them in a separate tuple and add them to a list.•**

```python
studens_information = list()
for key in students.keys():
    ad, soyad = key.split()
    studens_information.append((ad,soyad))

print(studens_information)
```

### **•4-Sort the names in alphabetical order and print the sorted list on the screen.•**

```python
print(sorted(students.keys()))
```

### **•5-Keep students with a GPA below 70 in a cluster (set).•**

```python
students_with_70_gpa = set()
for key, value in gpa_dict.items():
    if float(value) <= 70:
        students_with_70_gpa.add(key)
print(students_with_70_gpa)
```

## Question 2: Film Library Management System Project

•1-Create a movie data by taking information such as movie name, director, release year and genre from the user and store it in a dictionary.•**

•2-Give the user the option to edit or delete a movie. (For this, a suitable function must be written for whatever data they want to change about the movie.) 

•3-Allow the user to view their collection. List all movies or filter by criteria such as genre or year of release.•

•4-Save the movie data in a file and restore this data when you start the program.•


```python
from IPython.display import clear_output
movie_information = dict()



def koleksiyon_goruntuleme():
    clear_output()
    print("dosya okuma baslatiliyor ...\n")
    print("-"*30)
    dosya_okuma()
    print("dosya okuma basarili ...")

def dosya_yazma():
    with open("film_kutuphane.txt", "w") as dosya:
        for i,j in movie_information.items():
            dosya.writelines("{}\t{}\n".format(i,j))
        dosya.seek(0)  # Okuma yapabilmek için başa sar
        
def dosya_okuma():
    with open("film_kutuphane.txt", "r") as dosya:
        dosya.seek(0)
        print(dosya.read())

    # try:
    #     dosya = open("film_kutuphane.txt","a")
    #     if dosya.tell() != 0:
    #         dosya.seek(0)
    #         dosya.read()
    #     else:
    #         dosya.read()
    # except IOError:
    #     print("dosya okumada bir hata olurstu")
    # finally:
    #     dosya.close()

def film_ekleme():
    clear_output()
    film_adi = input("film adi giriniz: ").lower()
    film_yonetmen = input("yonetmen adi giriniz: ").lower()
    film_yili = int(input("film yilini giriniz: "))
    film_turu = input("film turunu giriniz: ").lower()
    
    movie_information[film_adi] = [film_yonetmen, film_yili, film_turu]
    dosya_yazma()

def film_duzenleme():
    clear_output()
    if "evet" == input("filmin adini degistirmek isiyormusun?(evet/hayir)?"):
        degistirilecek_filmin_adi = input("hangi filmin adini degistirmek istiyorsun?").lower()
        yeni_film_adi = input("filmin yeni adini giriniz ...").lower()

        if degistirilecek_filmin_adi in movie_information.keys():
            movie_information[yeni_film_adi] = movie_information.pop(degistirilecek_filmin_adi)
        else:
            print("Böyle bir film bulunamadı.")

        dosya_yazma()
        dosya_okuma()

    if "evet" == input("film detayini degistirmek istiyormusun?(evet/hayir)?"):

        degistirilecek_filmin_adi = input("hangi filmin detayini degistirmek istiyorsunuz?")
        kullanici_secenegi = input("hangi detayi degistirmek istiyorsun(yonetmen, film yili, film turu)").lower()

        if kullanici_secenegi == "yonetmen":
            movie_information[degistirilecek_filmin_adi][0] = input("yeni yonetmen adini giriniz: ")
        elif kullanici_secenegi == "filmyili" or "film yili":
            movie_information[degistirilecek_filmin_adi][1] = int(input("filmin yilini giriniz"))
        elif kullanici_secenegi == "filmturu" or "film turu":
            movie_information[degistirilecek_filmin_adi][2] = input("filmin turunu giriniz: ")
        
        dosya_yazma()
        dosya_okuma()

def film_silme():
    if "evet" == input("koleksiyonda film silmek istiyormusun(evet/hayir)?").lower():
            silinecek_film_adi = input("hangi filmi silmek istiyorsun?").lower()
            movie_information.pop(silinecek_film_adi)
            print("{:<3} film basarili bir sekilde silindi...".format(silinecek_film_adi))

            print("koleksiyonun son hali;\n")
            dosya_yazma()
            dosya_okuma()

    if "evet" == input("koleksiyondan film detayini silmek istiyormusun(evet/hayir)?"):

        silinece_film = input("hangi filmin detayini silmek isiyorsun")

        silinecek_film_detayi = input("filmin hangi detayi silmek istiyorsun(yonetmen, film yili, film turu)").lower()

        if silinecek_film_detayi == "yonetmen":
            movie_information[silinece_film].pop(0)
        elif silinecek_film_detayi == "film yili" or "film yili":
            movie_information[silinece_film].pop(1)
        elif silinecek_film_detayi == "film turu" or "filmturu":
            movie_information[silinece_film].pop(2)
        else:
            print("yanlis giris yapildi")

        dosya_yazma()
        dosya_okuma()





#-------------- algoritma baslangici -----------------------------

while True: 
    kullanici_girisi = int(input(""" yapmak istediginiz islemin numarasini giriniz(1-5): 
          1 - film ekleme
          2 - film duzenleme
          3 - film silme 
          4 - koleksiyon goruntuleme 
          5 - cikis yapma """))
    if kullanici_girisi == 1: 
        film_ekleme()  #en son dosyaya kayit islemi yapiliyor
    elif kullanici_girisi == 2: 
        film_duzenleme()  #en son dosyaya kayit islemi yapiliyor
    elif kullanici_girisi == 3: 
        film_silme()  #en son dosyaya kayit islemi yapiliyor
    elif kullanici_girisi == 4: 
        koleksiyon_goruntuleme()
    elif kullanici_girisi == 5: 
        break
    else:
        print("veri girisi yanlis")


```

## Question 3: Customer Management System
### Project Description: This project involves you creating a customer management system that you can use to manage your customers and perform basic transactions.
  This system will have basic functions such as storing customer information, adding new customers, updating customer information, deleting customers and viewing the customer list. 
 Here are the basic steps of the project:

 1-Use a dictionary structure to store customer information. Assign a unique customer identification (ID) for each customer and 
 associate customer information with this ID. You can use a dictionary containing information such as name, surname, e-mail, phone number for each customer.

 2-Provide a menu where the user can choose the following actions:

 •Add new customers
 •Updating customer information
 •Delete customer
 •List all customers
 •Check out

 3-Perform the relevant action according to the user's choice. For example, when adding a new customer, 
 get the required information from the user and add a new customer to the dictionary.

 4-When updating customer information, view the current information using the customer ID and save the updated information.

 5-In the customer deletion process, get the customer ID from the user and delete this customer from the dictionary.

 6-In the process of listing all customers, view the list of existing customers.

 7-Repeat the process until the user logs out.





```python
import random 
from IPython.display import clear_output

customer_info = dict()

def yeni_musteri_ekle():
    id = random.randint(10**4, 10**5 - 1)
    name = input("musteri adini giriniz: ") 
    surname = input("musteri soyadini giriniz: ") 
    eposta = input("musteri epostasini giriniz: ") 
    telno = int(input("musteri telno'sunu giriniz: "))
    customer_info[id] = [name,surname,eposta,telno]

def musteri_bilgilerini_guncelleme():
    customer_id = int(input("guncellemek istediginiz musterinin id'sini giriniz: "))
    option = input("hangi bilgiyi guncellemek istiyorsunuz(ad/soyad/e-posta/telefon numarasi)")
    if "ad" == option:
        customer_info[customer_id][0] = input("yeni adi giriniz").lower()
    elif "soyad" == option:
        customer_info[customer_id][1] = input("yeni soyadi girin").lower()
    elif "eposta" == option:
        customer_info[customer_id][2] = input("yeni eposta'yi giriniz")
    elif "telno" == option:
        customer_info[customer_id][3] = int(input("yeni telno giriniz: "))
    else:
        print("yanlis giris yapildi")

    print(customer_info)#mevcut durumu gosterme

def tum_musterileri_listeleme():
    clear_output()
    for i,j in customer_info.items():
        print("{} numarali musterinin bilgileri {}".format(i,j))

def musteri_silme():
    print(customer_info)
    customer_id = input("silmek istediginiz musterinin idsini giriniz")
    del customer_info[int(customer_id)]
    return print("silme islemi basarili...")





#-------------- algoritma baslangici -----------------------------

while True: 
    kullanici_girisi = int(input(""" yapmak istediginiz islemin numarasini giriniz: 
          1 - yeni musteri ekleme
          2 - musteri bilgilerini guncelleme
          3 - musteri silme 
          4 - tum musterileri listeleme 
          5 - cikis yapma """))
    if kullanici_girisi == 1: 
        yeni_musteri_ekle()
    elif kullanici_girisi == 2: 
        musteri_bilgilerini_guncelleme()
    elif kullanici_girisi == 3: 
        musteri_silme()
    elif kullanici_girisi == 4: 
        tum_musterileri_listeleme()
    elif kullanici_girisi == 5: 
        break
    else:
        print("veri girisi yanlis")


    
```
