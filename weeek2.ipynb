```python
###Question 2:
### All punts answering for Question num 2  in the code below 
#1-2-3-4##.save the movie library to a fileandrestorethisdata when the program starts.
import os

# File to store movie data (text format) :- TEXT FILE format 
MOVIES_FILE = "movies.txt"

def load_movies():
    """Load movies from text file if it exists."""
    movies = {}
    if os.path.exists(MOVIES_FILE):
        with open(MOVIES_FILE, 'r') as file:
            for line in file:
                # Split each line into components
                parts = line.strip().split('|')
                if len(parts) == 4:  # Ensure valid format
                    name, director, year, genre = parts
                    movies[name] = {
                        'director': director,
                        'year': year,
                        'genre': genre
                    }
    return movies

def save_movies(movies):
    """Save movies to text file in pipe-delimited format."""
    with open(MOVIES_FILE, 'w') as file:
        for name, details in movies.items():
            file.write(
                f"{name}|{details['director']}|"
                f"{details['year']}|{details['genre']}\n"
            )
#Added function to add:-
def add_movie(movies):
    """Add a new movie to the collection."""
    print("\n--- Add New Movie ---")
    name = input("Movie title (or 'exit' to cancel): ")
    if name.lower() == 'exit':
        return
    
    director = input("Director: ")
    year = input("Release year: ")
    genre = input("Genre: ")
    
    movies[name] = {
        'director': director,
        'year': year,
        'genre': genre
    }
    save_movies(movies)
    print(f"\n‚úÖ '{name}' added successfully!")
#Added function to edit :-
def edit_movie(movies):
    """Edit an existing movie's details."""
    print("\n--- Edit Movie ---")
    name = input("Enter movie title to edit: ")
    
    if name in movies:
        print(f"\nEditing: {name}")
        print(f"Current director: {movies[name]['director']}")
        movies[name]['director'] = input("New director (press Enter to keep): ") or movies[name]['director']
        
        print(f"\nCurrent year: {movies[name]['year']}")
        movies[name]['year'] = input("New year (press Enter to keep): ") or movies[name]['year']
        
        print(f"\nCurrent genre: {movies[name]['genre']}")
        movies[name]['genre'] = input("New genre (press Enter to keep): ") or movies[name]['genre']
        
        save_movies(movies)
        print(f"\n‚úÖ '{name}' updated successfully!")
    else:
        print(f"\n‚ùå Movie '{name}' not found!")
#added function to delete :-
def delete_movie(movies):
    """Remove a movie from the collection."""
    print("\n--- Delete Movie ---")
    name = input("Enter movie title to delete: ")
    
    if name in movies:
        del movies[name]
        save_movies(movies)
        print(f"\n‚úÖ '{name}' deleted successfully!")
    else:
        print(f"\n‚ùå Movie '{name}' not found!")
#Added function to display :-
def display_movies(movies, filter_type=None, filter_value=None):
    """Display movies with optional filtering."""
    print("\n--- Movie Collection ---")
    print("\n{:<30} {:<20} {:<10} {:<15}".format(
        "TITLE", "DIRECTOR", "YEAR", "GENRE"))
    print("-" * 75)
    
    if not movies:
        print("No movies in collection yet!")
        return
 ## filter movie by genre, year, or director:-   
    filtered_movies = {}
    if filter_type and filter_value:
        filter_value = filter_value.lower()
        if filter_type == 'genre':
            filtered_movies = {
                k: v for k, v in movies.items() 
                if v['genre'].lower() == filter_value
            }
        elif filter_type == 'year':
            filtered_movies = {
                k: v for k, v in movies.items() 
                if v['year'] == filter_value
            }
        elif filter_type == 'director':
            filtered_movies = {
                k: v for k, v in movies.items() 
                if v['director'].lower() == filter_value
            }
    
    display_list = filtered_movies if filtered_movies else movies
    
    for name, details in display_list.items():
        ''' print(f"\nüé¨ {name}")
        print(f"  Director: {details['director']}")
        print(f"  Year: {details['year']}")
        print(f"  Genre: {details['genre']}")'''
        print(" üé¨ {:<30} {:<20} {:<10} {:<15}".format(
            name[:28] + ('..' if len(name) > 28 else ''),
            details['director'][:18] + ('..' if len(details['director']) > 18 else ''),
            details['year'],
            details['genre'][:13] + ('..' if len(details['genre']) > 13 else '')))
    
    print(f"\nTotal movies: {len(display_list)}")
## Main function to run the program:-
def main():
    """Main program loop."""
    movies = load_movies()
    print(f"\nWelcome! Loaded {len(movies)} movies from storage.")
    
    while True:
        print("\n===== Movie Manager =====")
        print("1. Add Movie")
        print("2. Edit Movie")
        print("3. Delete Movie")
        print("4. View All Movies")
        print("5. Filter Movies")
        print("6. Exit")
        
        choice = input("\nChoose an option (1-6): ")
        
        if choice == '1':
            add_movie(movies)
        elif choice == '2':
            edit_movie(movies)
        elif choice == '3':
            delete_movie(movies)
        elif choice == '4':
            display_movies(movies)
        elif choice == '5':
            print("\nFilter by:")
            print("1. Genre")
            print("2. Year")
            print("3. Director")
            filter_choice = input("Choose filter type (1-3): ")
            
            if filter_choice == '1':
                genre = input("Enter genre to filter: ")
                display_movies(movies, 'genre', genre)
            elif filter_choice == '2':
                year = input("Enter year to filter: ")
                display_movies(movies, 'year', year)
            elif filter_choice == '3':
                director = input("Enter director to filter: ")
                display_movies(movies, 'director', director)
            else:
                print("Invalid filter choice!")
        elif choice == '6':
            print("\nSaving data... Goodbye!")
            break
        else:
            print("Invalid option! Please choose 1-6.")
## run the main function to start the program:-
if __name__ == "__main__":
    main()
```
