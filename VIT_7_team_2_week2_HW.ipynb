```python
### Question1: Student Grades Processing

You need to write a Python program to process a student's grades. The program needs to perform the following functions:

Store information and notes for 10 students using a dictionary. Let each student have their name, surname and grades (Midterm, Final and Oral grades). 
For example:

![image](https://github.com/user-attachments/assets/0e3f85a4-bf24-4b22-b0ea-f9b3dbda4fc8)

1-Calculate each student's GPA and add it to the dictionary.

2-Find the student with the highest GPA and print it on the screen.

3- Separate each student's name from their surname and store them in a separate tuple and add them to a list.

4-Sort the names in alphabetical order and print the sorted list on the screen.

5-Keep students with a GPA below 70 in a cluster (set).
```

```python
students={
    'Ahmet Yılmaz': [85,90,78], #Midterm, Final, Oral
    'Mehmet Demir': [92,88,76],
    'Ayşe Kaya': [78,89,95],
    'Zeynep Çelik': [65,70,80],
    'Ali Kara': [50,60,55],
    'Fatma Yıldız': [88,85,90],
    'Murat Aydın': [72,68,74],
    'Elif Aksoy': [95,90,88],
    'Hakan Öztürk': [45,50,55],
    'Canan Taş': [80,75,82] 
}

# 1-Calculate each student's GPA and add it to the dictionary.
for student in students:
    students[student].append(round((sum(students[student])/3),2))
print(students)
# 2-Find the student with the highest GPA and print it on the screen.
max_GPA=max(students.keys(), key=lambda x:students[x][-1])
print(f'Student with the highest GPA is: {max_GPA}\nHis/Her GPA is: {students[max_GPA][-1]}')

# 3- Separate each student's name from their surname and store them in a separate tuple and add them to a list.
names_surnames=[tuple(i.split()) for i in students]
print(names_surnames)

# 4-Sort the names in alphabetical order and print the sorted list on the screen.
for i in sorted(students.keys()):
    print(i)

# 5-Keep students with a GPA below 70 in a cluster (set).

GPA_below_70=dict(filter(lambda x: (x[1][-1])<70, students.items()))
GPA_below_70=set(GPA_below_70.keys())
print(f"GPA'i 70'in altında olan öğrenciler: {GPA_below_70}")
    
```

### Question 2: Film Library Management System Project

Project Description: This project aims to create an application that will help the user manage their movie collection. Users can add, edit, delete movies and view their collection.

Data Structures Used: Dictionaries (to store movies and related information), lists (to display movie collection)

Basic Functions:

1-Create a movie data by taking information such as movie name, director, release year and genre from the user and store it in a dictionary.

2-Give the user the option to edit or delete a movie. (For this, a suitable function must be written for whatever data they want to change about the movie.)

3-Allow the user to view their collection. List all movies or filter by criteria such as genre or year of release.

4-Save the movie data in a file and restore this data when you start the program.

```python
import json

# Veriyi dosyaya kaydeden fonksiyon
def save_data(movies):
    with open('movies.json', 'w') as file:
        json.dump(movies, file, indent=4)

# Veriyi dosyadan okuyan fonksiyon
def load_data():
    try:
        with open('movies.json', 'r') as file:
            return json.load(file)
    except FileNotFoundError:
        return []

# Film ekleme fonksiyonu
def create_movie():
    movies = load_data()

    if not movies:
        print("There has been no entry so far. So start from 1 as movie_ID")
    
    movie_ID = input('Enter Movie ID (Start from 1 and just go on...): ')
    movie_name = input('Enter movie name: ')
    director = input('Enter the director of the movie: ')
    release_year = input('Enter the release year: ')
    genre = input('Enter the genre of the movie: ')

    movie = {
        'movie_ID': movie_ID,
        'movie_name': movie_name,
        'director': director,
        'release_year': release_year,
        'genre': genre
    }

    movies.append(movie)
    save_data(movies)
    print("\nMovie added successfully! Current collection:")
    display_movies(movies)

# Film düzenleme fonksiyonu
def edit_movie():
    movies = load_data()
    if not movies:
        print("No movies to edit.")
        return

    display_movies(movies)
    choice = input('Enter the movie ID that you want to edit: ')

    for movie in movies:
        if movie['movie_ID'] == choice:
            print("Leave field empty if you don't want to change it.")
            new_name = input('New movie name: ')
            new_director = input('New director: ')
            new_year = input('New release year: ')
            new_genre = input('New genre: ')

            if new_name:
                movie['movie_name'] = new_name
            if new_director:
                movie['director'] = new_director
            if new_year:
                movie['release_year'] = new_year
            if new_genre:
                movie['genre'] = new_genre

            save_data(movies)
            print("Movie updated successfully!")
            display_movies(movies)
            return

    print("Movie ID not found.")

# Film silme fonksiyonu
def delete_movie():
    movies = load_data()
    if not movies:
        print("No movies to delete.")
        return

    display_movies(movies)
    choice = input('Enter the movie ID that you want to delete: ')

    updated_movies = [movie for movie in movies if movie['movie_ID'] != choice]

    if len(updated_movies) == len(movies):
        print("Movie ID not found.")
    else:
        save_data(updated_movies)
        print("Movie deleted successfully!")
        display_movies(updated_movies)

# Film görüntüleme ve filtreleme
def view_collection():
    movies = load_data()
    if not movies:
        print('There has been no entry so far.')
        return

    print('\nViewing Options:')
    print('1 - View all movies')
    print('2 - Filter by release year')
    print('3 - Filter by genre')
    choice = input('Enter your choice: ')

    if choice == '1':
        display_movies(movies)
    elif choice == '2':
        year = input('Enter the release year: ')
        filtered = [movie for movie in movies if movie['release_year'] == year]
        display_movies(filtered)
    elif choice == '3':
        genre = input('Enter the genre: ')
        filtered = [movie for movie in movies if movie['genre'].lower() == genre.lower()]
        display_movies(filtered)
    else:
        print("Invalid choice.")

# Yardımcı fonksiyon: Film listesini güzel şekilde yazdırır
def display_movies(movies):
    if not movies:
        print("No movies found.")
        return
    print("\n=== Movie Collection ===")
    for movie in movies:
        print(f"ID: {movie['movie_ID']} | Name: {movie['movie_name']} | "
              f"Director: {movie['director']} | Year: {movie['release_year']} | Genre: {movie['genre']}")
    print("========================\n")

# Ana menü döngüsü
while True:
    print("\n🎬 Film Library Management System")
    print("1. Add Movie")
    print("2. Edit Movie")
    print("3. Delete Movie")
    print("4. View Collection")
    print("5. Exit")
    choice = input("Enter your choice: ")

    if choice == '1':
        create_movie()
    elif choice == '2':
        edit_movie()
    elif choice == '3':
        delete_movie()
    elif choice == '4':
        view_collection()
    elif choice == '5':
        print("Exiting the system. Goodbye!")
        break
    else:
        print('Invalid choice! Please try again.')
```
